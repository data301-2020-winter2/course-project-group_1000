## Preliminaries

```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling as pdp
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv('Medical_Cost.csv')
```

```python
# Method chain
df = (
    pd.read_csv('Medical_Cost.csv')
    .rename(columns = {'age': 'Age', 'sex': 'Gender', 'bmi': 'BMI', 'children': 'Children', 'smoker': 'Smoker', 'region': 'Region', 'charges': 'Charges'})
    .round({'Charges': 2, 'BMI': 2})
    .replace({'yes': 'Yes', 'no': 'No'})
    .replace({'southwest': 'SW', 'southeast': 'SE', 'northeast': 'NE', 'northwest': 'NW'})
    .sort_values('Charges', ascending = True)
)
```

## 1. Distribution Plots

### 1.1 Age Distribution

```python
sns.displot(data = df, x = 'Age', bins = 25).set(title = "Age Distribution Plot")
```

***Observations:*** The distribution graph indicates that, age group of (18 ~ 20) accounts significantly more patients than the other age groups. Its patient count of close to 140 more than doubles the count on the tally of other ages. Apart from age 18~20, the dataset is consistent in the count of patients in each age bracket. 

### 1.2 Gender Distribution

```python
sns.displot(data = df, x = 'Gender').set(title = "Gender Distribution Plot")
```

***Observations:*** The gender distribution graph implies that the male and female patient count only differs by a small margin. Specifically, the dataset consists of 676 male patients and 662 female patients - with a small difference of 14 entries.

### 1.3 Children Distribution

```python
sns.displot(data = df, x = 'Children').set(title = "Patient with # of Children Distribution Plot")
```

***Observations:*** the patient count negatively correlates with the number of children in each family, that is, patients with 0 child significantly outnumbers the rest of the brackets and as the children count increases, the count of patients with the respective number of children decreases.

### 1.4 Smoker Distribution

```python
sns.displot(data = df, x = 'Smoker').set(title = "Smoker vs. non-smoker Plot")
```

***Observations:*** From the graph we are able to see that non-smokers outnumbers the number of smokers by a large margin. The dataset is comprised of 1064 non-smokers and 274 smokers.

### 1.5 Region Distribution

```python
sns.displot(data = df, x = 'Region').set(title = "Region Distribution Plot")
```

***Observations:*** We may conclude that the patients are evenly spread out across the four regions. 

### 1.6 Charge distribution

```python
sns.displot(data = df, x = 'Charges').set(title = "Medical Charges Distribution Plot")
```

***Observations:*** The charge distribution presents a right-skewed diagram that implies that most of the entries have medical charges below 20,000 dollars.

#### RQ1. People in which area have higher BMI, does it affect the average cost of insurance in the area?

***Analysis:*** From the box plot above, we can see that there is definitely a relationship between the regions in which patients are situated and their respective BMI values. Although the ranges along with the maximum and minimum values are relatively similar for all regions, we can see there is a slightly higher BMI medians in certain regions. The median for Northeast and Northwest regions are similar, as well as the interquartile ranges with not much difference. However, Southwest regions have a slightly higher median and a very minor increase in interquartile range, suggesting a slightly higher BMI amongst patients in that region. The most significant increase would be Southeast regions where the median, interquartile range, maximum and minimum values are noticeably higher than the other regions. Within these box plots, there are very few points outside the whiskers. There may be a few outliers however, they do not affect the general plot as there are a lot of entries in this dataset. The regression model beneath shows the relationship between BMI and the medical costs. Although there is no clear trend within the scatter plot, there is a rough upwards trend as shown on the trend line. This suggest that there may be a weak relationship between BMI and medical costs, suggesting that a higher BMI may generally lead to higher medical costs. However, there are still many patients that have low or healthy BMI that have high medical costs. This could be due to other factors other than BMI, such as diseases or illnesses that the patient may be undergoing treatment for which does not have any relation to BMI.
