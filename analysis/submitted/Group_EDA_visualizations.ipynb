## Preliminaries

```python
#import libraries
import numpy as np
import pandas as pd
import seaborn as sns
import pandas_profiling as pdp
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv('Medical_Cost.csv')
```

```python
# Method chain
df = (
    pd.read_csv('Medical_Cost.csv')
    .rename(columns = {'age': 'Age', 'sex': 'Gender', 'bmi': 'BMI', 'children': 'Children', 'smoker': 'Smoker', 'region': 'Region', 'charges': 'Charges'})
    .round({'Charges': 2, 'BMI': 2})
    .replace({'yes': 'Yes', 'no': 'No'})
    .replace({'southwest': 'SW', 'southeast': 'SE', 'northeast': 'NE', 'northwest': 'NW'})
    .sort_values('Charges', ascending = True)
)
```

## 1. Distribution Plots

### 1.1 Age Distribution

```python
sns.displot(data = df, x = 'Age', bins = 25)
```

***Observations:*** The distribution graph indicates that, age group of (18 ~ 20) accounts significantly more patients than the other age groups. Its patient count of close to 140 more than doubles the count on the tally of other ages. Apart from age 18~20, the dataset is consistent in the count of patients in each age bracket. 

### 1.2 Gender Distribution

```python
sns.displot(data = df, x = 'Gender')
```

***Observations:*** The gender distribution graph implies that the male and female patient count only differs by a small margin. Specifically, the dataset consists of 676 male patients and 662 female patients - with a small difference of 14 entries.

### 1.3 Children Distribution

```python
sns.displot(data = df, x = 'Children')
```

***Observations:*** the patient count negatively correlates with the number of children in each family, that is, patients with 0 child significantly outnumbers the rest of the brackets and as the children count increases, the count of patients with the respective number of children decreases.

### 1.4 Smoker Distribution

```python
sns.displot(data = df, x = 'Smoker')
```

***Observations:*** From the graph we are able to see that non-smokers outnumbers the number of smokers by a large margin. The dataset is comprised of 1064 non-smokers and 274 smokers.

### 1.5 Region Distribution

```python
sns.displot(data = df, x = 'Region')
```

***Observations:*** We may conclude that the patients are evenly spread out across the four regions. 

### 1.6 Charge distribution

```python
sns.displot(data = df, x = 'Charges')
```

***Observations:*** The charge distribution presents a right-skewed diagram that implies that most of the entries have medical charges below 20,000 dollars.

## 2. Age-Related Plots

#### 2.1 Age - BMI Relationship

```python
sns.lmplot(data = df, x = 'Age', y = 'BMI', palette="Set1")
```

***Observations:*** The regressional age-BMI plot indicates no clear correlation between the two attributes. It is apparent that entries are evenly scattered across the graph, thus we can conclude that imcreasing age does not heightens one's BMI.

#### 2.2 Age - BMI Relationship (With Gender)

```python
sns.lmplot(data = df, x = 'Age', y = 'BMI', hue = 'Gender', palette="Set1")
```

***Observations:*** Similar to graph 2.1, this graph depicts a clearer view that there are no correlations between Age and BMI even when gender are considered. From the graph, we see the two colors, (blue represents female patients, while red represents male patients) are evenly spreaded across the graph. There does not exist an area in the chart where either color are dominant, thus this affirms our observation.

#### 2.3 Age - BMI Relationship (With Region)

```python
sns.lmplot(data = df, x = 'Age', y = 'BMI', hue = 'Region', palette="Set1")
```

***Observations:*** Again, similar to the two charts above, entries are evenly spread across the graph and barely shows any relationship with the respective attributes.

#### 2.4 Age - children Relationship

```python
sns.stripplot(data = df, x = 'Children', y = 'Age', palette="pastel", jitter = True)
```

***Observations:*** From the chart we can see that there are no ties between the patient's age and the number of children they have. The reason why entries of families with 4 or 5 children are less saturated is due to lesser patient count as shown in section 1.3. Apart from that, there is no relationship with the age and children count.

## 7. Region-Related Plot

#### 7.1 Region - Charges  Relationship

```python
sns.pointplot(data = df, x = 'Region', y = 'Charges')
```

***Observations:*** The graph shows point estimates and the vertical line represents the confidence intervals. From here, we can clearly see the charge differences between the four regions with the south-east bound having noticeably higher medical charges. Interestingly, the cost in both western regions are about even and has shown to be cheaper than the eastern regions.
