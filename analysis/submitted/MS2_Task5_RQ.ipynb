#### RQ1. People in which area have higher BMI, does it affect the average cost of insurance in the area?

***Analysis:*** From the box plot above, we can see that there is definitely a relationship between the regions in which patients are situated and their respective BMI values. Although the ranges along with the maximum and minimum values are relatively similar for all regions, we can see there is a slightly higher BMI medians in certain regions. The median for Northeast and Northwest regions are similar, as well as the interquartile ranges with not much difference. However, Southwest regions have a slightly higher median and a very minor increase in interquartile range, suggesting a slightly higher BMI amongst patients in that region. The most significant increase would be Southeast regions where the median, interquartile range, maximum and minimum values are noticeably higher than the other regions. Within these box plots, there are very few points outside the whiskers. There may be a few outliers however, they do not affect the general plot as there are a lot of entries in this dataset. The regression model beneath shows the relationship between BMI and the medical costs. Although there is no clear trend within the scatter plot, there is a rough upwards trend as shown on the trend line. This suggest that there may be a weak relationship between BMI and medical costs, suggesting that a higher BMI may generally lead to higher medical costs. However, there are still many patients that have low or healthy BMI that have high medical costs. This could be due to other factors other than BMI, such as diseases or illnesses that the patient may be undergoing treatment for which does not have any relation to BMI.
